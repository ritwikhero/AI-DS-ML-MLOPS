**Question 5:** Write a Python program to calculate the factorial of a number.

```python
# Your code here
def factorial(n):
    if(n==0):
        return 1
    
    return n * factorial(n-1)

num = int(input("Enter a number"))
print("Factorial of", num, "is", factorial(num))
```
